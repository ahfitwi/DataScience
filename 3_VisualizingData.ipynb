## <font color='green'>5.1 Matplotlib</font>

Low level graphics library

Gives large amount of control over plots

Often times might be tedious

Similar to Matlab's graphical plotting library

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

%matplotlib inline
```

```python
x=np.linspace(0,np.pi*4, num=200)
```

```python
y=np.sin(x)
```

```python
plt.plot(x,y) # two cycles --> 4*pi
```

```python
# Clean graph using plt.show()
plt.plot(x,y)
plt.show()
```

```python
# X and Y labels
plt.plot(x,y)
plt.xlabel('x')
plt.ylabel('sin(x)')
plt.show()
```

```python
# Add legend
plt.plot(x,y, label='sin(x)')
plt.xlabel('x')
plt.ylabel('sin(x)')
plt.legend()
plt.show()
```

```python
# Change color
plt.plot(x,y, 'r',label='sin(x)')
plt.xlabel('x')
plt.ylabel('sin(x)')
plt.legend()
plt.show()
```

```python
z=np.cos(x)
```

```python
# More graphs of more than two functions
plt.plot(x,y, 'r',label='sin(x)')
plt.plot(x,z, 'b',label='cos(x)')
plt.xlabel('x')
plt.ylabel('f(x)')
plt.legend()
plt.show()
```

```python
# Set x limits
plt.plot(x,y, 'r',label='sin(x)')
plt.plot(x,z, 'b',label='cos(x)')
plt.xlabel('x')
plt.ylabel('f(x)')
plt.xlim([0,6])
plt.legend()
plt.show()
```

```python
# Subplotting
plt.figure(figsize=(10,5))
plt.subplot(1,2,1)
plt.plot(x,y,'r')

plt.subplot(1,2,2)
plt.plot(x,z,'b')

plt.tight_layout() #prevents overlapping
plt.show()
```

```python
# Object oriented version
fig=plt.figure()
axes=fig.add_axes([0.1,0.1,0.8,0.8]) # draw axes
axes.plot(x,y,label='sin(x)')
axes.legend()
plt.show()
```

## <font color='green'>5.2 Graphs with Matplotlib</font>

How to produce histograms, scatter, 2D rectangular, and hexagonal histograms using Matplotlib.

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

%matplotlib
```

```python
df=pd.read_csv('auto-mpg.data',sep='\s+')
cols=['mpg','cylinders','dis','hp','weight','acc','year','origin','name']
df.columns=cols
df.head()
```

```python
#Histogram
plt.hist(df['mpg'])
plt.show()
```

```python
#Increase the numbe rof bins
plt.figure(figsize=(10,6))
plt.hist(df['mpg'],bins=20)
plt.xlabel('mpg')
plt.show()
```

```python
plt.figure(figsize=(10,6))
plt.scatter(x=df['weight'],y=df['mpg'],alpha=0.5) # transparency
plt.xlabel('weight')
plt.ylabel('mpg')
plt.show()
```

```python
plt.figure(figsize=(10,6))
plt.hist2d(x=df['weight'],y=df['mpg'])
plt.xlabel('weight')
plt.ylabel('mpg')
plt.colorbar()
plt.show()
```

```python
plt.figure(figsize=(10,6))
plt.hexbin(x=df['weight'],y=df['mpg'],gridsize=(35,15))
plt.xlabel('weight')
plt.ylabel('mpg')
plt.colorbar()
plt.show()
```

## <font color='green'>5.3 Graphs with Seaborn</font>

How to produce histogram, scatter, bar, and boxplots using Seaborn.

Seaborn:

    Based on Matplotlib
    High-level plotting
    Easier to work with compared to Matplotlib

```python
import pandas as pd
import numpy as np
import matplotlib as plt
import seaborn as sns
import warnings

warnings.filterwarnings('ignore')
%matplotlib inline
```

```python
df=pd.read_csv('auto-mpg.data',sep='\s+')
cols=['mpg','cylinders','dis','hp','weight','acc','year','origin','name']
df.columns=cols
df.head()
```

```python
sns.distplot(df['mpg'])
```

```python
#plt.figure(figsize=[10,6])
sns.distplot(df['mpg'],kde=False, bins=20)
#plt.show()
```

```python
sns.lmplot('weight','mpg',data=df)
```

```python
sns.lmplot('weight','mpg',data=df,fit_reg=False)
```

```python
sns.lmplot('weight','mpg',data=df,fit_reg=False,aspect=2, size=6)
```

```python
sns.lmplot('weight','mpg',data=df,fit_reg=False,aspect=2, size=6,hue='cylinders')
```

```python
sns.lmplot('weight','mpg',data=df,fit_reg=False,aspect=1, size=4,hue='cylinders',col='origin')
```

```python
sns.kdeplot(df['weight'],df['mpg'],shade=True)
```

```python
sns.jointplot('weight','mpg',data=df,alpha=0.4)
```

```python
g=sns.PairGrid(df,diag_sharey=False)
g.map_lower(sns.kdeplot)
#g.map_upper(plt.scatter)
g.map_diag(sns.kdeplot)
```

```python
sns.countplot(y='cylinders',data=df)
```

```python
sns.boxplot(x='cylinders',y='mpg',data=df)
```

```python
sns.boxplot(x='origin',y='mpg',data=df,hue='cylinders')
```

## <font color='green'>5.4 Graphs with Pandas</font>

Directly produce graphs by using pandas plotting methods

Quick and easy (sometimes dirty) solution

```python
import pandas as pd
import numpy as np
import matplotlib as plt
import seaborn as sns
import warnings

warnings.filterwarnings('ignore')
%matplotlib inline
```

```python
df=pd.read_csv('auto-mpg.data',sep='\s+')
cols=['mpg','cylinders','dis','hp','weight','acc','year','origin','name']
df.columns=cols
df.head()
```

```python
df['weight'].plot(kind='hist',bins=20)
```

```python
df.groupby('cylinders')['mpg'].plot(kind='hist')
```

```python
df.groupby('cylinders')['mpg'].plot(kind='hist',alpha=0.4,legend=True)
#plt.show()
```

```python
df.plot(kind='scatter', x='weight',y='mpg')
```

```python
df.boxplot(column='mpg',by='cylinders',grid=False)
```

```python
df.pivot_table(index='cylinders',aggfunc='mean')
```

```python
df.pivot_table(index='cylinders',aggfunc='sum')
```

```python
dfp=df.pivot_table(index='cylinders',aggfunc='mean')
```

```python
dfp.columns
```

```python
dfp.plot.line(y='acc')
```

```python
for col in dfp.columns:
    dfp.plot.line(y=col)    
```

```python
df.groupby('cylinders').count()
```

```python
df.groupby('cylinders').count()['mpg'].plot(kind='bar')
```
