```python
df5.head()
```

```python
df5.iloc[1:4] # row 1 to row 3, integer based indexing
```

```python
df5.iloc[1:4, [3,4]] # row 1 to row 3 and columns 3 and 4
```

## <font color='green'>4.5 Boolean Indexing in DataFrames</font> 

Perform filtering on rows based on Boolean indexing

The Boolean array is obtained by comparing a variable with a value, similar to the way it was done in Numpy

```python
import pandas as pd
import seaborn as sns
```

```python
df5.head()
```

```python
df5['c']>310
```

```python
(df5['c']>310) & (df5['f']>11) # use and or or (& and |)
```

```python
df5[(df5['c']>310) & (df5['f']>11)]#subsetting
```

```python
df5.loc[df5['a']==15,['a','g','h']]
```
