<font color='red'>Iris Datset</font>

Four columns:sepal length (cm)', 'sepal width (cm)', 'petal length (cm)', 'petal width (cm)'

```python
from sklearn import datasets
dat=datasets.load_iris()
dat.feature_names
```

```python
dat.target
```

```python
len(dat.target)
```

```python
dat.data.shape
```

```python
import pandas as pd
df=pd.DataFrame(dat.data,columns=dat.feature_names)
```

```python
df['target']=dat.target
df1=df
```

```python
df.head(5)
```

```python
# Adding the color types
```

```python
# Using a function coupled iwth apply
def addColor(x):
    if x==0:
        return "red"
    elif x==1:
        return "green"
    else:
        return "blue"
```

```python
df['color']=df['target'].apply(addColor)
```

```python
df.head()
```

```python
# Using function, lambda, and apply
df2=df1
df2['color']=df2['target'].apply(lambda x: addColor(x))
```

```python
df2.tail(3)
```

```python
# Using lambda and apply only
df1['color']=df1['target'].apply(lambda x:"red" if x==0 else ("green" if x==1 else "blue"))
```

```python
df1.tail(4)
```

```python
df1.to_csv('iris.csv')
```
