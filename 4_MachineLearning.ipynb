### <font color='green'>1.1: Machine Learning</font> 

Machine Learning (ML): a subfield of AI that involves self-learning algorithms where a model can learn from data without being explicitly programmed. The model will then be able to predict outcomes of new data (generalization).

"A computer program is said to learn experience E with respect to some class of tasks T and performance measure P, id its performance at tasks T, as measured by P, improves with experience E." -Tom M.Mitchell

Used in many areas: Fraud Detection, Credit approval, Text sentiment analysis, Speech recognition, image recognition, Email spam filtering, Service Recommendation.

### <font color='green'>1.2: Types of Machine Learning</font> 

Types of Machine Learning are are of three broad types:
    
    Supervised
    Unsupervised
    Reinforcement

Supervised Learning: Target variable present for training. For example: Regression, Classification (X,y)

Unsupervised Learning: No target variable present. For example: CLustering, Dimensionality reduction (X)

Reinforcement Learning: Reward oriented learning algorithms based on past actions and results.

$X_{1}$, $X_{2}$, ..., $X_{p}$: Independent Variable, Input, Feature, Predictor

Y: Dependent Variable, Outcome, Response

### <font color='green'>Reinforcement Learning</font> 

Agent interacts with environment and improves its behavior based on feedback/results from the past.

Action - Reaction

Application areas generally involve agent based environments where some sort of reward is present such as games, agent based simulation etc

Differs from supervised learning in that no direct input-output data is present. The outcome might be obtained several steps later which may be in the form of success/failure. The agent then analyzes its past behavior and updates its priorities and plays the game again. This will lead the agent to become better over time.

Trial and error learning method

The output vector can be N by 1 DataFrame or N element pandas series or 1D Numpy array.
